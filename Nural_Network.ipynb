#### Importing necessary libraries:

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
# To convert categorical labels to numerical ones using "one hot" encoding
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder as LBE                                
```

#### Defining network structure variables:

```python
number_of_neurons = [50, 100]
learning_rate = [0.07, 0.5]
batch_size = [64, 128]
repeat = 50
```

#### Reading-in data:

```python
import scipy.io
data = scipy.io.loadmat('binaryalphadigs')
# print(data)
print("Primary data type: ", type(data))
print("Keys in dict: ", data.keys(), "\n")

table = data['dat']
print("Main data shape: ", table.shape)
# print(table)
labels = data['classlabels']
print("Shape of labels: ", labels.shape, "\n")

label_array = [i[0][0] for i in labels.T]
print("Labels: ", label_array)
num_classes = len(label_array)
print("Number of classes (The same as number of labels): ", num_classes, "\n")

print("Dimension of a picture: ", table[0][0].shape)
input_dimension = table[0][0].shape[0] * table[0][0].shape[1]
print("Expected size of input array by flattening a picture array: ", input_dimension)
```

#### Pre-processing:

```python
# Converting each picture array to a 1-dimension array:
table_to_1d = [[np.matrix.flatten(table[i][j]) for j in range(len(table[i]))] for i in range(len(table))]
print("Picture arrays' dimensions after flattening: ", table_to_1d[0][0].shape, "\n")
# print(table_to_1d)

# Making a 1-D array of all data, instead of a 36*39 matrix:
table_to_1row = []
for i in range(len(table_to_1d)):
    for j in range(len(table_to_1d[i])):
        table_to_1row.append(table_to_1d[i][j])
X = np.array(table_to_1row)
print("Main data shape after converting it to a 1-row array: ", X.shape, "\n")

# Making an appropriate label array:
# print(type(label_array))
temp_labels = np.array(label_array).T
Y = []
for i in range(36):
    for j in range(39):
        Y.append(temp_labels[i])
Y = np.array(Y)
Y = LBE().fit_transform(Y)  # Convert to numerical array
Y = to_categorical(Y, num_classes)  # Encode to one-hot 36 ta kelas driii be ezaye har kelas yedone araye 36 binary miszi moadel ba on kelas to hamon yedone khune 1  36 dar 1 ke khune aval sefr 1 has 
print("Final labels shape: ",Y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y)
```

```python
# One-layered NN:
for n in number_of_neurons:
    for lr in learning_rate:
        for bt_sz in batch_size:
            model = Sequential()
            model.add(Dense(n, input_shape=(input_dimension,), activation='sigmoid'))
            model.add(Dense(num_classes))
            model.summary()
            model.compile(optimizer=SGD(lr=lr), loss='mse', metrics=['accuracy'])
            model.fit(x_train, y_train, batch_size=bt_sz, epochs=repeat)
```

```python
# Two-layered NN:
for i in number_of_neurons:
    for j in number_of_neurons:
        for lr in learning_rate:
            for bt_sz in batch_size:
                model = Sequential()
                model.add(Dense(i, input_shape=(input_dimension,), activation='sigmoid'))
                model.add(Dense(j, activation='sigmoid'))
                model.add(Dense(num_classes))
                model.summary()
                model.compile(optimizer=SGD(lr=lr), loss='mse', metrics=['accuracy'])
                model.fit(X, Y, batch_size=bt_sz, epochs=repeat)
```
